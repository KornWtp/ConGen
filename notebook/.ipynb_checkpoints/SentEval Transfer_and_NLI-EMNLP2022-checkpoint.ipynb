```python
import pandas as pd
from glob import glob
import numpy as np
```

```python
def mean_all(dataframe):
    all_item = dataframe.drop(columns=['model']).values.tolist()
    results = dataframe.drop(columns=['model']).columns
    for idx,key in enumerate(results):
        avg = 0
        for _,all_x in enumerate(all_item):
            for i, x in enumerate(all_x):
                if i == idx:
                    avg+=x
        print(key)
        print(round(avg/len(all_item),2))
    print(f"Avg.:{round(np.mean(all_item),2)}")
```

```python
task = 'nli'
file = glob(f"exp_report/*_{task}*.csv")
for f in file:
    print(f)
    df_score = pd.read_csv(f).sort_values(by='avg', ascending=False)
    mean_all(df_score)
    print('*'*50)
    print()
```

```python
task = 'clf'
file = glob(f"exp_report/*_{task}.csv")
for f in file:
    print(f)
    df_score = pd.read_csv(f).sort_values(by='avg', ascending=False)
    mean_all(df_score)
    print('*'*50)
    print()
```
